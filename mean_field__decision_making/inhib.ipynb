```python
def default_params():
    """
    Returns a dictionary of default parameters for the neural dynamics simulation.
    """
    params = {
        # Simulation timing and duration
        "time_steps": 1000,  # Total duration in timesteps
        "dt": 0.002,  # Timestep size in seconds

        # Timescales for population dynamics (in seconds)
        "tau_NMDA": 0.1,  # NMDA (excitatory) population timescale
        "tau_GABA": 0.005,  # GABA (inhibitory) population timescale
        "tau_AMPA": 0.002,  # AMPA (background) timescale
        "tau_In": 0.002,  # Input stimulus timescale

        # Neural population parameters
        "a": 2,  # Gain of the excitatory population
        "b": 10,  # Threshold of the excitatory population
        "d": 0.15,  # Slope of the excitatory population

        # Connectivity strengths
        "w_values": {
            "w_EE": [0],  # Excitatory-to-excitatory connectivity
            "w_EI": [0, 0.2, 0.5],  # Excitatory-to-inhibitory connectivity
            "w_IE": [0],  # Inhibitory-to-excitatory specificity
        },

        # Stimulation parameters
        "stimulus_strength": 20,  # Intensity of stimulus
        "coherence_range": (0.5, 1),  # Range for coherence bias
        "stimulation_period": (0.4, 0.6),  # Fraction of time allocated to stimulation

        # Noise parameters
        "noise_mean": 0.0,  # Mean of input noise
        "noise_std": 0.2,  # Standard deviation of input noise

        # Initial conditions
        "initial_rates": {
            "E1": 0,  # Initial rate for excitatory population E1
            "E2": 0,  # Initial rate for excitatory population E2
            "I1": 0,  # Initial rate for inhibitory population I1
            "I2": 0,  # Initial rate for inhibitory population I2
            "AMPA": 0,  # Initial rate for AMPA inputs
        },
        "init_ranges": {
            "E1": (0, 0),  # Range for initial conditions of E1
            "E2": (0, 0),  # Range for initial conditions of E2
            "I1": (0, 0),  # Range for initial conditions of I1
            "I2": (0, 0),  # Range for initial conditions of I2
        },

        # Validation thresholds
        "valid_tresh": 15.0,  # Threshold for valid excitatory activity
        "valid_low_tresh": 5.0,  # Low threshold for excitatory activity

        # Simulation settings
        "r_init_n": 10,  # Number of initial conditions for excitatory populations
        "r_init_n2": 10,  # Number of initial conditions for inhibitory populations
        "n_trials": 10,  # Number of trials to simulate
        "tolerance": 1e-4,  # Tolerance for numerical computations
    }
    return params
```
